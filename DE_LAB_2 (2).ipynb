```python
import pandas as pd
import numpy as np
import io

from IPython.display import display
from ipywidgets import FileUpload

# Create uploader widget
uploader = FileUpload(accept='.csv', multiple=False)
display(uploader)
```

```python
# Read uploaded file
if len(uploader.value) == 0:
    print("Please upload the Iris.csv file")
else:
    for filename, fileinfo in uploader.value.items():
        content = fileinfo['content']
        df = pd.read_csv(io.BytesIO(content))

    print("Iris dataset loaded successfully.")
    display(df.head())
```

```python
filtered_df = df[df['PetalLengthCm']>5]
filtered_df.head()
```

```python
pivot_table = pd.pivot_table(
    df,
    values='PetalLengthCm',
    index='Species',
    aggfunc='mean'
)
#display pivot table
pivot_table
```

```python
melted_df= pd.melt(
    df,
    id_vars=['Species'],
    value_vars=['SepalLengthCm', 'SepalWidthCm'],
    var_name='Feature',
    value_name='Value'
)
melted_df.head();
```

```python
duplicate_count = df.duplicated().sum()
print("no. of duplicate rows",duplicate_count)
```

```python

duplicates_count= df.duplicated().sum()
print("Number of duplicates rows:", duplicates_count)
```

```python
sepal_length_array = np.array(df['SepalLengthCm'])
mean_value = np.mean(sepal_length_array)
std_deviation =  np.std(sepal_length_array)
variance = np.var(sepal_length_array)

print("mean of sepal length",mean_value)
print("standard deviation of sepal length", std_deviation)
print("variance of sepal length", variance)
```

```python
species_info= pd.DataFrame({
    'Species': ['Iris-setosa','Iris-versicolor','Iris-virginica'],
    'Category':['Small','Medium','Large']
})

merged_df= pd.merge(df,species_info,on='Species')

merged_df.head()
```
